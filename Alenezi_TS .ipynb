The following graph represents the speed against the character name

```python
plotdata = pd.DataFrame(
    {"Name": pokemon["Name"],
                   "Speed" : pokemon["Speed"]})
plotdata.set_index('Name').plot(kind = "line", grid=True)
```

The following graph represents the Attack against the character name

```python
plotdata = pd.DataFrame(
    {"Name": pokemon["Name"],
                   "Attack" : pokemon["Attack"]})
plotdata.set_index('Name').plot(kind = "line", grid=True)
```

How the speed effect the attacking power of the character?

The following graph represents the Attacking value, Speed value against the character name.
When we have plot both the attributes in a single graph, It provides us the following trend.
From the trend, it is analyzed that with increasing the value of the speed of the pokemon, the attacking power also increased.

```python
plotdata = pd.DataFrame(
    {"Name": pokemon["Name"],
                   "Speed" : pokemon["Speed"],
                   "Attack" : pokemon["Attack"]})
plotdata.set_index('Name').plot(kind = "line", grid=True)
```

How many number of characters are legendary?

The graph that will clearly explain that how many number of characters are legendary and how many number of characters are not legendary.

```python
legendaryValues = pokemon["Legendary"].value_counts()
plt.title("Precentage legendary and not legendary characters")
legendaryValues.plot(kind = "bar", fontsize = 14)
plt.show()
legendaryValues.plot(kind = "pie", autopct = "%1.0f%%" )
```

What is the relation between the legendary characters and the attacking power?

It is clear from the graph that when the attacking power is less then the most of the characters are not legendary but when the attacking power is greater then most of the characters are legendary.

```python
#superheroes: crosstab of FIRST_YEAR & alignment ALIGN
pokemonLegendary = pd.crosstab(pokemon['Speed'], pokemon["Legendary"])
#print(appearAlign)
pokemonLegendary.plot()
plt.title("Number of Legendary character against the Attacking Power")
plt.ylabel("Number of Characters", fontsize = 12)
plt.xlabel("Attacking Power", fontsize = 14)
#distinguish where graph goes with plt.show
plt.show()
```
